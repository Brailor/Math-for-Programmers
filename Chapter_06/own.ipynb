```python
class QuadraticFunction2(Vector):
   def __init__(self,a,b,c):
       self.a = a
       self.b = b
       self.c = c
   def add(self,v):
       return QuadraticFunction2(self.a + v.a, 
                                self.b + v.b, 
                                self.c + v.c)
   def scale(self,scalar):
       return QuadraticFunction2(scalar * self.a, 
                                scalar * self.b, 
                                scalar * self.c)
   def __call__(self,x):
       return self.a * x * x + self.b * x + self.c
   @classmethod
   def zero(cls):
       return QuadraticFunction2(0,0,0)
```

```python
f = QuadraticFunciton(1, 1, 1)

plot([f], -11, 10)
```
